# Black-Scholes Baseline

We extract the implied vol smile from live SPX options data.
The point is simple: Black-Scholes assumes constant vol.
Markets don't. The gap between those two things is what this whole project is about.

```python
import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt
from data.loader import load_spx_options
from models.black_scholes import implied_vol, compute_smile
```

```python
r = 0.05
spot, expiry, T, strikes, prices = load_spx_options(expiry_index=4)
```

```python
strikes, ivols = compute_smile(strikes, prices, spot, T, r)

atm_idx = np.argmin(np.abs(strikes - spot))
atm_vol = ivols[atm_idx]
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(strikes, ivols * 100, 'o-', color='steelblue', linewidth=2, label='Implied vol (market)')
ax.axhline(y=atm_vol * 100, color='tomato', linestyle='--', linewidth=1.5, label=f'BS flat vol (ATM = {atm_vol*100:.1f}%)')
ax.axvline(x=spot, color='gray', linestyle=':', alpha=0.6, label='ATM spot')

ax.set_xlabel('Strike')
ax.set_ylabel('Implied Volatility (%)')
ax.set_title(f'SPX Volatility Smile  |  Expiry {expiry}  |  Spot {spot:.0f}')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../results/01_bs_smile.png', dpi=150)
plt.show()

print(f"\nATM vol:  {atm_vol*100:.2f}%")
print(f"Smile range: {ivols.min()*100:.2f}% — {ivols.max()*100:.2f}%")
```

The red dashed line is what Black-Scholes predicts: the same vol at every strike.
The blue curve is what the market actually prices — lower vol for OTM calls, higher for OTM puts.

That downward skew (sometimes called the vol smirk on SPX) is the target both Heston and Rough Bergomi are built to reproduce.
